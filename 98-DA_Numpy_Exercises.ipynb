# Exercice Numpy

## 1. Array creation

- Create a 1D array called ```xarray``` with values from 0 to 10 and in steps of 0.1. Check the shape of the array:

````python
xarray = np.arange(0, 10, 0.1)
xarray.shape
````

- Create an array of normally distributed numbers with mean $\mu=0$ and standard deviation $\sigma=0.5$. It should have 20 rows and as many columns as there are elements in ```xarray```. Call it ```normal_array```:

````python
normal_array = np.random.normal(0, 0.5, (20, xarray.size))
normal_array
````

- Check the type of ```normal_array```:

````python
normal_array.dtype
````

## 2. Array mathematics

- Using ```xarray``` as x-variable, create a new array ```yarray``` as y-variable using the function $y = 10* cos(x) * e^{-0.1x}$:

````python
yarray = 10 * np.cos(xarray) * np.exp(-0.1 * xarray)
yarray
````

- Create ```array_abs``` by taking the absolute value of ```yarray```:

````python
array_abs = np.abs(yarray)
array_abs
````

- Create a boolean array (logical array) where all positions $>0.3$ in ```array_abs``` are ```True``` and the others ```False```

````python
array_abs > 0.3
````

- Create a standard deviation projection along the second dimension (columns) of ```normal_array```. Check that the dimensions are the ones you expected. Also are the values around the value you expect?

````python
std_projection = normal_array.std(axis=1)
# std_projection.shape
std_projection
````

## 3. Plotting

- Use a line plot to plot ```yarray``` vs ```xarray```:

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_title('xarray vs yarray')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(xarray, yarray)
````

- Try to change the color of the plot to red and to have markers on top of the line as squares:

````python
fig, ax = plt.subplots()
ax.set_title('xarray vs yarray')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(xarray, yarray, 's-r')
````

- Plot the ```normal_array``` as an image and change the colormap to 'gray':

````python
fig, ax = plt.subplots()
ax.imshow(normal_array, cmap='gray')
````

- Assemble the two above plots in a figure with one row and two columns grid:

````python
fig, ax = plt.subplots(1,2)
ax[0].plot(xarray, yarray, 's-r')
ax[1].imshow(normal_array, cmap = 'gray')
````

## 4. Indexing

- Create new arrays where you select every second element from xarray and yarray. Plot them on top of ```xarray``` and ```yarray```.

````python
xarray_2 = xarray[::2]
yarray_2 = yarray[::2]

fig, ax = plt.subplots()
ax.scatter(xarray, yarray)
ax.scatter(xarray_2, yarray_2, marker='x')
````

- Select all values of ```yarray``` that are larger than 0. Plot those on top of the regular ```xarray``` and ```yarray```plot.

````python
pos_idx = yarray > 0
pos_x = xarray[pos_idx]
pos_y = yarray[pos_idx]

fig, ax = plt.subplots()
ax.scatter(xarray, yarray)
ax.scatter(pos_x, pos_y, marker='x', c = 'r')
````

- Flip the order of ```xarray``` use it to plot ```yarray```:

````python
fig, ax = plt.subplots()
ax.scatter(xarray[::-1], yarray)
````

## 5. Combining arrays

- Create an array filled with ones with the same shape as ```normal_array```. Concatenate it to ```normal_array``` along the first dimensions and plot the result:

````python
normal_ones = np.ones_like(normal_array)
# normal_ones

normal_plus_one = np.concatenate((normal_ones, normal_array), axis=0)
plt.imshow(normal_plus_one)
````
